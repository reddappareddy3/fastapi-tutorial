#### HTTP Methods

HTTP Methods (also called HTTP verbs) specify the type of action a client wants to perform on a resource available on a server.

They are part of the HTTP protocol used in:
- Web applications
- REST APIs
- Clientâ€“server communication

#### Why HTTP Methods are Used

HTTP methods help to:
- Define clear communication between client and server
- Follow RESTful architecture
- Improve security and scalability
- Standardize API operations

## Commonly Used HTTP Methods

##  1. GET

### Definition:

Used to **retrieve data** from the server.

### Features:

* Does **not modify** server data
* Data sent through **URL**
* Safe and idempotent
* Can be cached

### Example:

```http
GET /products/5
```

### Use Cases:

* Fetch user details
* Display products
* Load web pages

- Not used for sending sensitive data



##  2. POST

### Definition:

Used to **send data to the server** to create a new resource.

### Features:

* Data sent in **request body**
* Not idempotent
* Cannot be cached
* More secure than GET

### Example:

```http
POST /users
```

### Use Cases:

* User registration
* Login
* Form submission



## 3. PUT

### Definition:

Used to **update an existing resource completely**.

### Features:

* Replaces the entire resource
* Idempotent
* Requires full data

### Example:

```http
PUT /users/10
```

### Use Cases:

* Update full profile
* Replace existing data


##  4. PATCH

### Definition:

Used to **partially update a resource**.

### Features:

* Updates specific fields only
* More efficient than PUT
* May not be idempotent

### Example:

```http
PATCH /users/10
```

### Use Cases:

* Update email
* Change password
* Modify single column


##  5. DELETE

### Definition:

Used to **remove a resource** from the server.

### Features:

* Deletes data
* Idempotent
* Usually no request body

### Example:

```http
DELETE /users/10
```
